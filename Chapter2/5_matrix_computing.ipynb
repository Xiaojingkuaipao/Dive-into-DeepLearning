#### 标量对一个向量求导，求导结束之后向量会转置， 列向量会变成一个行向量
$$
x = \begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
\quad \Rightarrow \quad
\frac{dy}{dx} = \begin{bmatrix}
\frac{dy}{dx_1} & \frac{dy}{dx_2} & \cdots & \frac{dy}{dx_n}
\end{bmatrix}
$$

**公式样例，要记住**
- 表1
  - | $y$              | $a$             | $au$            | $sum(x)$        | $‖x‖²$           |
      |----------------|---------------|---------------|---------------|----------------|
      | $\frac{dy}{dx}$ | $0^T$      | $a \frac{du}{dx}$ | $1^T$      | $2x^T$       |

- 表2
  - | y               | $ u + v $              | $ uv $                | $ \langle u, v \rangle $(表示内积)  |
    |-----------------|--------------------------|-------------------------|-----------------------------|
    | $ \frac{dy}{dx} $ | $ \frac{du}{dx} + \frac{dv}{dx} $ | $ \frac{du}{dx} v + \frac{dv}{dx} u $ | $ u^T \frac{dv}{dx} + v^T \frac{du}{dx} $ |

#### 如果是**向量**对**标量**求导，求完导向量不转置
$$
y = \begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_m
\end{bmatrix}
\quad \Rightarrow \quad
\frac{\partial y}{\partial x} = \begin{bmatrix}
\frac{dy_1}{dx} \\
\frac{dy_2}{dx} \\
\vdots \\
\frac{dy_m}{dx}
\end{bmatrix}
$$

#### 如果是向量对向量求导
- if $X \in R^n, y \in R^m, \Rightarrow \frac{\partial y}{\partial x} \in R^{m \times n}$


<img src="images/5_image_2.png" height="300" alt="Matrix Derivative">

$$
\begin{array}{|c|c|c|c|c|}
\hline
y & a & x & Ax & x^T A \\
\hline
\frac{dy}{dx} & 0 & I & A & A^T \\
\hline
\end{array}
$$
$$
\begin{array}{|c|c|c|c|}
\hline
y & au & Au & u + v \\
\hline
\frac{\partial y}{\partial x} & a \frac{\partial u}{\partial x} & A \frac{\partial u}{\partial x} & \frac{\partial u}{\partial x} + \frac{\partial v}{\partial x} \\
\hline
\end{array}
$$
